# Dibujar un diagrama

### El procedimiento mÃ­nimo para dibujar un diagrama de Venn es el siguiente.

> https://recursospython.com/guias-y-manuales/diagramas-de-venn-con-matplotlib/

```python
from matplotlib import pyplot as plt
from matplotlib_venn import venn2
```

```python
venn2((4, 2, 1))
```

#### Argumentos: (a, b, c) : a = tamaÃ±o Diagrama A. b = tamaÃ±o Diagrama B. c = tamaÃ±o Interseccion A y B.
#### En el ejemplo a = 4, b = 2, c = 1. a == Diagrama A. b== Diagrama B. c == Interseccion A y B.

> El argumento indica el tamaÃ±o relativo de los tres subconjuntos en este orden: 
**Ab (izquierda)**, **aB (derecha)**, **AB (intersecciÃ³n).** 
AsÃ­, la tupla (4, 2, 1) dibujarÃ­a el conjunto A el doble de tamaÃ±o respecto de B.

> Para identificar a cada uno de los subconjuntos (tres en diagramas de dos conjuntos) el mÃ³dulo utiliza una nomenclatura que consiste en colocar un 1 para indicar que la secciÃ³n estÃ¡ incluida en el conjunto y un 0 para indicar que estÃ¡ excluida. De esta manera, siguiendo el orden **Â«ABCÂ»**, el subconjunto 10 es el de la izquierda (el que pertenece a A pero no a B); el 01, el de la derecha (el que pertenece a B pero no a A); y el 11, el del medio (la intersecciÃ³n).

```python
# set_colors determina el color de los conjuntos.

venn2((1, 1, 1), set_colors=("#888888", "#0000FF"))

# El color de las instersecciones es calculado automÃ¡ticamente.
```

```python
# NÃ³tese que, por defecto, tienen un valor alpha de 0.4, i.e., 60% de transparencia.
# UtilÃ­cense valores de 0.0 a 1.0 en el parÃ¡metro de ese mismo nombre.
# Deshabilita la transparencia.

venn2((1, 1, 1), set_colors=("#888888", "#0000FF"), alpha=1)
```

```python
# La funciÃ³n venn2() retorna una instancia de una clase llamada VennDiagram.
# En particular, los mÃ©todos de Ã©sta que nos interesan son get_label_by_id() y get_patch_by_id().
# Ambas toman un subconjunto y retornan instancias de matplotlib.text.Text y matplotlib.patches.PathPatch, respectivamente.
# A partir de ellas podemos modificar individualmente el aspecto de cada subconjunto.
# Por ejemplo, el siguiente cÃ³digo cambia el color y el texto del subconjunto 11 (i.e. AB).

diagrama_venn = venn2((4, 2, 1))
diagrama_venn.get_label_by_id("11").set_text("IntersecciÃ³n")
diagrama_venn.get_label_by_id("11").set_color("#FFFFFF")
```

```python
# O bien alterar el color de un subconjunto en particular: PARA CALCULAR LA MEZCLA de COLORES,
# como la formacion apartir de los colores primarios


diagrama_venn = venn2((4, 2, 1))
diagrama_venn.get_label_by_id("11").set_text("IntersecciÃ³n")
diagrama_venn.get_label_by_id("11").set_color("#FFFFFF")

# Establece el color de cada Diagrama A y B.
diagrama_venn.get_patch_by_id("10").set_color('y') # yellow : amarillo
diagrama_venn.get_patch_by_id("01").set_color('b') # blue : azul

# Establece el color de la intersecciÃ³n:
diagrama_venn.get_patch_by_id("11").set_color('g') # green : verde (inteserccion entre amarillo y azul)
```

```python
# El parÃ¡metro set_labels permite cambiar los nombres de los conjuntos mostrados en el diagrama.

venn2((1, 1, 1), set_labels=("ROJO", "AZUL"))
```
