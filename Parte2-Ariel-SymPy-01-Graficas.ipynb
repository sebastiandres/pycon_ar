## Graficas en 3D

> La funcion **plot3d ()** genera un grafico tridimensional. **plot3d (expr, xrange, yrange, kwarg s)** 

```python
from sympy.plotting import plot3d
```

```python
from sympy import *
```

```python
x = Symbol('x')
```

```python
y = Symbol('y')
```

```python
x, y 
```

```python
x
```

```python
y
```

### El siguiente ejemplo dibuja un grafico de superficie 3D

```python
plot3d (x * y, (x, -10,10), (y, -10,10))
```

```python
plot3d (x * y, x / y, (x, -5, 5), (y, -5, 5))
```

```python
plot3d(x*y, -x*y, (x, -5, 5), (y, -5, 5))
```

```python
 plot3d((x**2 + y**2, (x, -5, 5), (y, -5, 5)),
...     (x*y, (x, -3, 3), (y, -3, 3)))
```

> La funcion **plot3d_parametric_line ()** representa un grafico lineal parametrico en **3 dimensiones.**

```python
from sympy.plotting import plot3d_parametric_line
```

```python
plot3d_parametric_line (cos (x), sin (x), x, (x, -5, 5))
```

 ### Para dibujar una grafica de superficie parametrica, use la funcion plot3d_parametric_surface ().

> plot3d_parametric_surface (xexpr, yexpr, zexpr, rangex, rangey, kwargs)

```python
##https://www.cfm.brown.edu/people/dobrush/am33/SymPy/part1.html
```

```python
from sympy import symbols, cos, sin
from sympy.plotting import plot3d_parametric_surface
u, v = symbols('u v')
```

```python
plot3d_parametric_surface(cos(u + v), sin(u - v), u - v,(u, -5, 5), (v, -5, 5))
```

```python
from sympy import plot_implicit, cos, sin, symbols, Eq, And
x, y = symbols('x y')
```

```python
# Using mesh grid with number of points as input.
```

```python
p5 = plot_implicit(Eq(x**2 + y**2, 5),
...         (x, -5, 5), (y, -2, 2),
...         adaptive=False, points=400)
```

```python
# Plotting regions.
```

```python
p6 = plot_implicit(y > x**2)
```

```python
# Plotting Using boolean conjunctions.
```

```python
p7 = plot_implicit(And(y > x, y > -x))
```
